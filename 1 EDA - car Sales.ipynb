```python
# comparing max sale data with minimum
def min_max_values(col):
    
    top_index = df[col].idxmax()
    top_obs = pd.DataFrame(df.loc[top_index])
    
    bot_index = df[col].idxmin()
    bot_obs = pd.DataFrame(df.loc[bot_index])
                           
    min_max_obs =pd.concat([bot_obs,top_obs],axis = 1)
    
    return min_max_obs
```
